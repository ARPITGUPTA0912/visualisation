Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x ="timepoint" and y = "signal" for different events and regions.

```python
df1= sns.load_dataset("fmri")
```

```python
df1.head()
```

```python
sns.lineplot(data = df1,x ="timepoint" ,y = 'signal' ,hue ='event' , style ='region')
```

Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.

```python
df2=sns.load_dataset("titanic")
```

```python
df2.head()
```

```python
import matplotlib.pyplot as plt
sns.boxplot(data= df2, x='pclass', y ='age' )
plt.title("titanic Dataset 1")
```

```python
sns.boxplot(data= df2, x='pclass', y ='fare' )
plt.title("titanic Dataset 2")
```

Que 6: Use the "flights" dataset from seaborn to plot a heatmap.


```python
df5= sns.load_dataset('flights')
```

```python
df5.head()
```

```python
flights= df5.pivot_table(index="month", columns="year", values="passengers")
sns.heatmap(flights, annot=True,fmt='d')
```
